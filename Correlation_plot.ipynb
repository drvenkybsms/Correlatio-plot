```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme(style="white")

# generate sample correlation matrix
#df = sns.load_dataset("titanic")
#df["label"] = np.random.randint(0, 2, size=len(df))
corr = df1.corr()

# mask and melt correlation matrix
mask = np.tril(np.ones_like(corr, dtype=bool)) | corr.abs().le(0.001)
melt = corr.mask(mask).melt(ignore_index=False).reset_index()
melt["size"] = melt["value"].abs()

fig, ax = plt.subplots(figsize=(8, 6))



# normalize colorbar
cmap = plt.cm.RdYlBu
norm = plt.Normalize(-1, 1)
sm = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
cbar = plt.colorbar(sm, ax=ax, aspect = 10, shrink=0.4, pad=0.005, anchor=(0,1))  # Adjusted pad and anchor parameters
cbar.ax.tick_params(labelsize="x-small")





# plot lower triangle (scatter plot with normalized hue and circle markers)
sns.scatterplot(ax=ax, data=melt, x="index", y="variable",
                hue="value", hue_norm=norm, palette=cmap,
                style=0, markers=["o"], legend=False)

# Increase the size of the markers by adjusting the s parameter
ax.scatter(melt["index"], melt["variable"], s=melt["size"]*300, c=melt["value"], cmap=cmap, vmin=-1, vmax=1)



# get the diagonal elements of the correlation matrix
diagonal_correlations = np.diag(corr)

# plot diagonal with color squares
for i, corr_value in enumerate(diagonal_correlations):
    color = sm.to_rgba(corr_value)
    plt.scatter(i, i, s=300, c=[color], marker="o")


# format grid
xmin, xmax = (-0.5, corr.shape[0] - 0.5)
ymin, ymax = (-0.5, corr.shape[1] - 0.5)
ax.vlines(np.arange(xmin, xmax + 1), ymin, ymax, lw=1, color="silver")
ax.hlines(np.arange(ymin, ymax + 1), xmin, xmax, lw=1, color="silver")
ax.set(aspect=1, xlim=(xmin, xmax), ylim=(ymax, ymin), xlabel="", ylabel="")
ax.tick_params(labelbottom=True, labeltop=False)
plt.xticks(rotation=45)



# ... (the below code is only to fix the rotated label at the particular column)

# Assuming 'columns' is a list of column names in your DataFrame
columns = df1.columns

# Set ticks for x-axis (bottom)
tick_positions = np.arange(len(columns))  # Use 0-based tick positions
ax.set_xticks(tick_positions)
ax.set_xticklabels(columns, rotation_mode='anchor', ha='right', fontsize=9)

# Set ticks for y-axis (left)
ax.set_yticks(tick_positions)
ax.set_yticklabels(columns, rotation=45, rotation_mode='anchor', va='center', fontsize=9)






# annotate upper triangle
for y in range(corr.shape[0]):
    for x in range(corr.shape[1]):
        value = corr.mask(mask).to_numpy()[y, x]
        if pd.notna(value):
            plt.text(x, y, f"{value:.2f}", size="x-small",
                     # color=sm.to_rgba(value), weight="bold",
                     ha="center", va="center")
            
            
# Title of the plot
plt.title('')

# Save the plot in high resolution to a specific folder
output_folder = 'E://Dr.Pradeepa//eGFR-Paper//Revised_200'  # Replace with your folder path 
output_filename = 'Correlation.png'
output_path = f'{output_folder}/{output_filename}'
plt.savefig(output_path, dpi=600, bbox_inches='tight')  # Adjust dpi for desired resolution
```
